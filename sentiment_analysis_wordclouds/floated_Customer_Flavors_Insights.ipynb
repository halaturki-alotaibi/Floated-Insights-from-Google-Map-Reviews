```python
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# قراءة البيانات
df = pd.read_csv("processed_reviews (1).csv")

# قائمة النكهات المحتملة
flavors = ["كراميل مكياتو", "فول سوداني و مربى", "مناقيش", "تشيز كيك", "كريم بروليه", "كلاسيك","كوكو بيري","لقيمات جبن","قرفة و بيكان","كليجا","تمر و طحينة","بوينو","أومبا لومبا","فستق مملح","بنانا بودنق","تيراميسو","يوزو و نعناع"]

# عد التكرارات
flavor_counts = Counter()

for clean_text in df["clean_text"].dropna():
    for flavor in flavors:
        if flavor in clean_text:
            flavor_counts[flavor] += 1

# إنشاء WordCloud باستخدام قاموس التكرارات
wc = WordCloud(
    font_path="/content/Amiri-Bold.ttf",
    width=800,
    height=400,
    background_color="white"
).generate_from_frequencies(flavor_counts)

# عرض النتيجة
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
